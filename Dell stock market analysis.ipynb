```python
q1 = data['Volume'].quantile(0.25)
q3 = data['Volume'].quantile(0.75)

iqr = q3-q1
upper_bound = q3+1.5*iqr
lower_bound = q1-1.5*iqr
```

```python
data = data[~(data>upper_bound)|(data<lower_bound)]
```

```python
 def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
x = data.drop(['Open'] , axis = 1)
y = data['Open']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 , random_state = 42)
```

```python
x_train.shape , x_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(x_train , y_train)
```

```python
model_pred = model.predict(x_test)
```

```python
model_pred
```

```python
pd.DataFrame({'Actual':y_test , 'Predicted': model_pred})
```

```python
mse = np.mean((y_test-model_pred)**2)
```
